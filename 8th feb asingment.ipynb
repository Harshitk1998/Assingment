```python
# ANSWER 3)

# The abc module in Python stands for "Abstract Base Classes". 
# It is used to define abstract base classes for Python classes. 
# Abstract base classes are classes that can't be instantiated, but they define a common API that can be used by subclasses.

# The abc module provides the ABC class that is used to define abstract base classes. 
# To create an abstract base class, you need to subclass ABC and define abstract methods in it.
# Abstract methods are methods that don't have any implementation in the abstract base class. Subclasses must provide their own implementation for these methods.

# The abc module is used to enforce a certain structure and behavior in the classes that inherit from the abstract base classes. 
# It helps in designing better interfaces, providing clarity in code and reducing errors.
```
