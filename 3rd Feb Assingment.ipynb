```python
# Answer 4)
# A generator is a special type of function which does not return a single value, instead, it returns an iterator object with a sequence of values. It allows you to generate a series of values on the fly, rather than generating them all at once and storing them in memory.
# The yield keyword is used to return a value from the generator function and pause the execution of the function until the next value is requested. 
 
def generator(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```

```python
a = generator(5)
```

```python
for i in a:
    print(i)
```

```python
# Answer 6)
def fibonacci():
    a,b = 0,1
    while True:
        yield a
        a,b = b,a+b
```

```python
fib = fibonacci()
```

```python
for i in range(10):
    print(next(fib))
```
